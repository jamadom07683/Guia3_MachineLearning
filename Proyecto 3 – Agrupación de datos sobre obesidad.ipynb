## Planteamiento de Hipótesis

A continuación, se presentan dos hipótesis iniciales relacionadas con los patrones de obesidad y su relación con los hábitos y características demográficas de los individuos:

#### Visualización de Datos y Reducción de Dimensionalidad
##### 
Usa técnicas como PCA (Análisis de Componentes Principales) para reducir la dimensionalidad y visualizar los datos:

```python
# Selección de variables cuantitativas
X = df[variables_numericas]

# Reducción de dimensionalidad usando PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# kmeans ajustado 
kmeans = KMeans(n_clusters=10, random_state=42)  # Ajusta el número de clusters según tu análisis
kmeans.fit(X)
labels = kmeans.labels_

# Visualización de los datos en el plano cartesiano
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='viridis')
plt.title('Clusters')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.show()

```

##### Determinación de los Clusteres usando el método del codo

```python
# Fijar una semilla
np.random.seed(42)

# Selección de variables cuantitativas
X = df[variables_numericas]

# Determinación del número de grupos usando el método del codo
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)


plt.plot(range(1, 11), wcss)
plt.title('Método del Codo')
plt.xlabel('Número de Clusters')
plt.ylabel('WCSS')
plt.show()
```

Después de analizar los resultados de los diferentes modelos de agrupamiento, concluimos que el algoritmo K-means con 3 clústeres proporcionó la mejor separación de los datos. Esto se debe a que el método del codo indicó un número óptimo de clústeres en 3, por que la variación es menos drastica y las visualizaciones PCA mostraron una clara separación de grupos. Además, al comparar la inercia y las visualizaciones, K-means demostró ser el más eficaz en mantener la compactación y la separación de los clústeres.

```python
# Generar datos
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Aplicar K-means
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Graficar los puntos y los centroides
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75)
plt.title('Agrupamiento K-means')
plt.xlabel('Característica 1')
plt.ylabel('Característica 2')
plt.show()
```

#### Conclusiones

Después de aplicar el algoritmo de agrupamiento K-Means y analizar la separación de los clusters con técnicas como PCA, determinamos que el modelo con 3 clusters proporciona la mejor organización de los datos. El método del codo indicó que el punto óptimo de agrupación ocurre con 3 clusters, ya que la variación en la inercia comienza a estabilizarse después de este punto. Además, las visualizaciones de PCA mostraron una clara separación de los grupos, lo cual confirma la cohesión interna de cada cluster y su separación respecto a los demás.

Comparado con otros números de clusters, K-Means con 3 clusters fue el más eficaz para mantener la compactación y separación, destacando patrones significativos en los datos. Esto permite realizar una interpretación más clara de los niveles de obesidad en función de las variables cuantitativas seleccionadas, ofreciendo una base sólida para futuras decisiones basadas en estos grupos.
